#Machine learning
###Machine Learning (ML) is a subset of artificial intelligence that enables computers to learn from data and make predictions or decisions without being explicitly programmed. It involves algorithms and statistical models that identify patterns and relationships in data, making it valuable for tasks like image recognition, natural language processing, and predictive analytics.

#Supervised Learning Algorithm:
###Supervised learning is a machine learning paradigm in which an algorithm learns from labeled data to make predictions or decisions. It involves two main phases: training and testing. During training, the algorithm learns to map input data to corresponding output labels. Then, during testing, it uses this knowledge to make predictions on new, unseen data.


###Real-Life Example:

####Imagine you're building a spam email filter (a classification task). You start by collecting a dataset of emails, where each email is labeled as either "spam" or "not spam." In this dataset, the emails are the input data (features), and their corresponding labels (spam or not spam) are the output.

###1. Training Phase

  * You feed the algorithm this labeled dataset during training.
  * The algorithm learns patterns and relationships between the content and attributes of emails (features) and their respective labels (spam or not spam).


###2. Testing Phase

  * Once trained, the algorithm is deployed to analyze incoming emails.
  * It uses the knowledge gained during training to predict whether a new email is spam or not spam based on its content and attributes.

##Over-fitting & Under-fitting

### **Overfitting:** Overfitting occurs when a machine learning model is excessively complex and learns to capture noise or random fluctuations in the training data. It fits the training data extremely closely, including the noise, which makes the model perform poorly on unseen data. In overfitting, the model has high variance, meaning it is sensitive to small changes in the training data, leading to poor generalization.

image.png

### **Underfitting:** Underfitting happens when a machine learning model is too simplistic to capture the underlying patterns in the data. It lacks the capacity or complexity to represent the data adequately. This results in a model that performs poorly not only on the training data but also on new, unseen data. In underfitting, the model has high bias, as it makes strong assumptions about the data that do not hold true.

#Linear Regression

###Linear Regression is a supervised learning algorithm used for predicting a continuous target variable (also known as the dependent variable) based on one or more input features (independent variables). It assumes a linear relationship between the input features and the target variable.

### **Simple Linear Regression:**
####If a single independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Simple Linear Regression.
### **Multiple Linear regression:**
####If more than one independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Multiple Linear Regression.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset (car mileage vs. price)
data = pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/vehicle_dataset.csv").head(10)

# Extract input feature (mileage) and target variable (price)
X = data["Mileage (km/l)"].values.reshape(-1, 1)
Y = data['Price (Rupees)']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Make a prediction for a new mileage value
#We reshape it to have a single feature (column) using .reshape(-1, 1)
#because scikit-learn expects input features in a specific shape.


new_mileage =np.array([40]).reshape(-1,1)
predicted_price = model.predict(new_mileage)

plt.scatter(X, Y,color='g')
plt.plot(X, Y,color='k')
plt.scatter( new_mileage,predicted_price, color='b')
plt.xlabel("Mileage (KM/l)")
plt.ylabel("Price (in Rupees)")
plt.show()


print(f"Predicted Price for {new_mileage[0][0]}K Miles: ${predicted_price[0]:.2f}")
```

#Naive Bayes
###Naive Bayes is a probabilistic machine learning algorithm commonly used for classification tasks. It's based on Bayes' theorem and the assumption of independence between features (hence the term "naive"). Naive Bayes calculates the probability of a data point belonging to each class and selects the class with the highest probability as the prediction.

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np
import pandas as pd

# Sample dataset: Fruit weight in grams and their corresponding labels (0 for apple, 1 for orange)
data =pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/fruit_data.csv")

# Separate features (weights) and labels
X =data['Weight'].values.reshape(-1, 1)
Y = data['Label']

# Create a Naive Bayes classifier
classifier = GaussianNB()

# Train the classifier on the data
classifier.fit(X, Y)

# New fruit weight to classify

new_weight = 140

# Predict whether the new fruit is an apple or an orange
predicted_label = classifier.predict([[new_weight]])

if predicted_label[0] == 0:
    print("It's an apple!")
else:
    print("It's an orange!")
```

# Decision Tree
###A Decision Tree is a popular machine learning algorithm used for both classification and regression tasks. It works by recursively splitting the data into subsets based on the most significant attribute at each step, resulting in a tree-like structure where each internal node represents a decision based on an attribute, and each leaf node represents a class label or a numerical value.

###(Regression predicts continuous numerical values, like predicting house prices. Classification predicts discrete categories, like classifying emails as spam or not.)

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Sample dataset: Loan approval prediction
# Features: Credit Score and Income, Labels: Approved (1) or Rejected (0)
data = pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/loan_approval_dataset.csv")

# Separate features (credit score and income) and labels (approved or rejected)
X = data[['Credit_Score', 'Annual_Income']]
Y = data['Approval_Status']

# Create a Decision Tree classifier
classifier = DecisionTreeClassifier()

# Train the classifier on the data
classifier.fit(X, Y)

# New loan application to predict
new_application = [[650, 50000]]

# Predict whether the new application is approved or rejected
predicted_label = classifier.predict(new_application)


if predicted_label[0] == 1:
    print("Loan application approved!")
else:
    print("Loan application rejected.")
```

#Random Forest
###Random Forest is an ensemble learning algorithm that combines multiple decision trees to improve predictive accuracy and reduce overfitting. It creates a "forest" of decision trees, each trained on a random subset of the data with bootstrapping (bagging), and each tree's output contributes to the final prediction. It's effective for both classification and regression tasks.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Sample dataset: Customer data for product purchase prediction
# Features: Age and Income, Labels: Purchase (1) or No Purchase (0)
data =pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/customer_data.csv")

# Separate features (age and income) and labels (purchase or no purchase)
X = data[["Age","Income"]]
Y = data["Label"]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a Random Forest classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42) #"n_estimators=100"  =  100 decision trees

# Train the classifier on the training data
classifier.fit(X_train, Y_train)


# Make predictions on the test data
Y_pred = classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
```
