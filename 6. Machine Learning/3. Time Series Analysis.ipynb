#Time Series Analysis
### **Time Series Analysis** is a statistical technique used to analyze and extract meaningful insights from time-ordered data. It involves examining patterns, trends, and dependencies within a dataset to make predictions or uncover underlying behaviors. Time series data is commonly encountered in various fields, such as finance, economics, climate science, and sales forecasting.

###( **'Value':** This key is associated with an array of values generated using NumPy **(np)** . It creates a sine wave using **np.sin** with an increasing range of values from 0 to 99, scaled by 2Ï€ and divided by 30 to create a repeating pattern. Noise is added to this pattern by adding random values drawn from a normal distribution using **np.random.randn.** )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample time series data
df=pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/infy_stock.csv").head(50)

# Create a DataFrame from the data
print(df["Date"])
print(df["High"])


# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['High'], label='Time Series Data')
plt.xlabel('High')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.title('Time Series Analysis Example')
plt.legend()
plt.grid(True)
plt.show()
```

##Forcasting

###**Forecasting** is a statistical and data analysis technique used to make predictions about future values or events based on historical data and patterns. It's widely used in various fields, including finance, economics, sales, weather forecasting, and more. Forecasting helps businesses and researchers plan for the future by estimating what might happen.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample time series data
df=pd.read_csv("/content/drive/MyDrive/Data Science/6. Machine Learning/Dataset/infy_stock.csv").head(50)

# Create a DataFrame from the data
print(df["Date"])
print(df["High"])




# Calculate the 7-day moving average
df['Moving_Average'] = df['High'].rolling(window=7).mean()
# Explanation:

# - .rolling(window=7) creates a rolling window of size 7, which means it considers the last 7 rows at a time.
# - .mean() computes the mean (average) of the values within the rolling window.
# For example, a 7-day moving average computes the average of the past 7 days'
#worth of data at each point in the time series.

# Plot the original time series data and the moving average
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['High'], label='Time Series Data')
plt.plot(df['Date'], df['Moving_Average'], label='7-Day Moving Average', linestyle='--')
plt.xlabel('Date')
plt.xticks(rotation=90, ha="right")
plt.ylabel('High')
plt.title('Time Series Forecasting Example (Moving Average)')
plt.legend()
plt.grid(True)
plt.show()
```

###linestyle

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Plot with different linestyles
plt.plot(x, y, linestyle='-', label='Solid Line')      # Solid line
plt.plot(x, y, linestyle='--', label='Dashed Line')    # Dashed line
plt.plot(x, y, linestyle=':', label='Dotted Line')     # Dotted line
plt.plot(x, y, linestyle='-.', label='Dash-Dot Line')  # Dash-Dot line

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Style Examples')
plt.legend()
plt.grid(True)
plt.show()
```
