#Dataset for Skewness and Kurtosis Analysis

###In this example, we will work with a small dataset of 15 numbers to understand how to calculate skewness and kurtosis. Skewness measures the degree of asymmetry in a data distribution, while kurtosis quantifies the tailedness or peakedness of the distribution compared to a normal distribution.

```python
12, 15, 18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42
```

###Skewness Calculation:

Calculate the mean: Sum all the numbers and divide by the total count.
Mean = (12 + 15 + 18 + ... + 42) / 15 = 27

Calculate the median: Since there's an odd number of data points, the median is the middle value, which is 30.

Calculate the standard deviation: You can use a calculator or software to find this.

Use the skewness formula:
Skewness = (3 * (Mean - Median)) / Standard Deviation

```python
import numpy as np

# Given dataset
data = np.array([12, 15, 18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42])

# Calculate mean and median
mean = np.mean(data)
median = np.median(data)

# Calculate standard deviation
std_dev = np.std(data)

# Calculate skewness using the formula
skewness = (3 * (mean - median)) / std_dev

print(f"Skewness: {skewness}")
```

```python
from scipy.stats import skew

data1 = np.array([12, 15, 18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42])

skew_data=skew(data1)
print(skew_data)
```

Kurtosis Calculation:

Calculate the fourth power deviations from the mean for each data point.

Sum up these fourth power deviations.

Calculate the standard deviation of the dataset.

Use the kurtosis formula:
Kurtosis = (Sum of ((Xi - Mean)^4) / n) / (Standard Deviation ^ 4)

```python
import numpy as np

# Given dataset
data = np.array([12, 15, 18, 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42])

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate fourth power deviations from the mean
fourth_power_deviations = np.power(data - mean, 4)

# Sum up the fourth power deviations
sum_fourth_power_deviations = np.sum(fourth_power_deviations)

# Calculate kurtosis using the formula
kurtosis_value = (sum_fourth_power_deviations / len(data)) / (std_dev ** 4)

print(f"Kurtosis: {kurtosis_value}")
```

```python
from scipy.stats import kurtosis

data1 = np.array([ 20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42])
kurt=kurtosis(data1)
print(kurt)
```
