#Exploratory Data Analysis (EDA):
Exploratory Data Analysis is a crucial step in data analysis that involves summarizing, visualizing, and understanding the main characteristics of a dataset. It helps you gain insights, identify patterns, and formulate hypotheses for further analysis.

##Data Types:
Data types describe the nature of the values stored in a dataset. Common data types include:

###Numeric:
Representing quantitative values. Examples include integers and floating-point numbers.
###Categorical:
Representing categories or labels. Examples include strings, enums, or boolean values.
###Ordinal:
A categorical type with an inherent order, such as education levels (e.g., high school, college, Ph.D.).
###Datetime:
Representing dates and times.
###Text:
Representing textual data.

##Four Levels of Data:
Data is often categorized into four levels based on its nature and characteristics:

###Nominal:
Categorical data with no inherent order (e.g., colors, genders).
###Ordinal:
Categorical data with an ordered relationship (e.g., education levels).
###Interval:
Numeric data with consistent intervals but no true zero (e.g., temperature in Celsius).
###Ratio:
Numeric data with consistent intervals and a true zero point (e.g., height, weight).

##Descriptive Statistics:

Descriptive statistics involve numerical and graphical summaries that describe the main features of a dataset. Common measures include:

###Mean:
Average value of the data.
###Median:
Middle value when data is sorted.
###Mode:
Most frequently occurring value.
###Standard Deviation:
Measure of data's spread around the mean.
###Min/Max:
Minimum and maximum values.
###Percentiles:
Values below which a given percentage of observations fall.

#Mean

The mean value is the average value.

To calculate the mean, find the sum of all values, and divide the sum by the number of values:

image.png

```python
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)

print(x)
```

#Median

The median value is the value in the middle, after you have sorted all the values:

image.png

```python
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

#Distribution

###Mesokurtic:

Imagine a hill or a bump that represents the data values. A mesokurtic distribution is like a normal hill – not too flat, not too pointy. It's kind of balanced, like the hills you often see. This means that the data has tails (values far from the center) that aren't too extreme or rare.

###Leptokurtic:

Now, picture a hill that's really tall and skinny, almost like a tower. A leptokurtic distribution has a higher peak than a normal hill and very thin tails that stretch out far. This means that the data has values that are concentrated closely around the center, and extreme values are more likely than in a normal distribution.

###Platykurtic:

Imagine a hill that's not very tall and has a wide, flat top. A platykurtic distribution has a lower peak than a normal hill and tails that spread out widely. This means that the data has values that are more spread out from the center, and extreme values are less likely than in a normal distribution.


So, in simple terms:

**Mesokurtic** is balanced and moderate.

**Leptokurtic** is tall and skinny, with extreme values.

**Platykurtic** is short and flat, with less extreme values.

image.png

```python
import pandas as pd
from scipy.stats import kurtosis

# Read the CSV file into a DataFrame
data = pd.read_csv("/content/drive/MyDrive/Data Science/EDA/health_data.csv")

# Calculate the kurtosis of the Max Pulse column
max_pulse_kurtosis = kurtosis(data['Max_Pulse'])

# Classify the distribution based on kurtosis value
if max_pulse_kurtosis < 0:
    distribution_type = "Platykurtic"
elif max_pulse_kurtosis > 0:
    distribution_type = "Leptokurtic"
else:
    distribution_type = "Mesokurtic"

print(f"The Max Pulse distribution is: {distribution_type}")
print(f"Kurtosis value: {max_pulse_kurtosis}")
```

**Here are code examples for each of the points mentioned in the overview of Exploratory Data Analysis (EDA):**

```python
import numpy as np
import pandas as pd


# Example data
numeric_data = [35,41,42,56,58,62,70,71,77,90]
print("Numeric Data:", numeric_data)


# Descriptive Statistics
mean_value = np.mean(numeric_data)
median_value = np.median(numeric_data)
std_deviation = np.std(numeric_data)
std_deviation = np.var(numeric_data)
min_value = np.min(numeric_data)
max_value = np.max(numeric_data)
percentile_25 = np.percentile(numeric_data, 40)
percentile_75 = np.percentile(numeric_data, 75)

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
print("Min Value:", min_value)
print("Max Value:", max_value)
print("50th Percentile:", percentile_25)
print("75th Percentile:", percentile_75)


```
