##Binomial Distribution
###Binomial distribution: counts the number of successful outcomes in a fixed number of repeated yes/no experiments with a consistent probability.

####Example:- A coin is thrown 5 times what is the probability obtaining,

1.   4 Heads
2.   1 Head

```python
from scipy.stats import binom

# Define the parameters
n = 5  # Number of trials (coin tosses)
p = 0.5  # Probability of success (getting a head)

# Calculate the probability of obtaining 4 heads
k = 4  # Number of successful outcomes (heads)
probability = binom.pmf(k, n, p)


print(f'Probability of getting 4 heads in 5 tosses: {probability:.4f}')
```

```python
from scipy.stats import binom

# Define the parameters
n = 7  # Number of trials (coin tosses)
p = 0.5  # Probability of success (getting a head)

# Calculate the probability of obtaining 4 heads
k = 5  # Number of successful outcomes (heads)
probability = binom.pmf(k, n, p)

print(f'Probability of getting 4 heads in 5 tosses: {probability:.2f}')
```

##Normal Distribution

###A normal distribution is a way to describe how numbers tend to be distributed around an average value in a predictable, bell-shaped pattern. It's used in many areas of science and statistics because it's a common way that things naturally happen.

###Example:- Assume that mean weight of students in a pharmacy institute to be 65KG with a variance of 16 kg. how many students in the Institute of 400 students would you expect to be over 72KG?

```python
mean_weight = 65
variance = 16
threshold_weight = 72
std = int(variance**0.5)

z_score = (threshold_weight - mean_weight) / std
print(z_score)

from scipy.stats import norm
probability_less_than_72 = norm.cdf(z_score)
print(probability_less_than_72)

# Calculate the probability of a student scoring over 72
probability_over_72 = 1 - probability_less_than_72

# Calculate the expected number of students scoring over 72
total_students = 400
expected_students_over_72 = round(probability_over_72 * total_students)

print(f"Expected number of students scoring over 72: {expected_students_over_72}")
```

###Example:- Assume that mean age of students in a college to be 70% with standard deviation of 5% how many students in the Institute of 60 students would you expect to be scored over 80% ?

```python
mean_age = 70
std = 5
threshold_value = 80

z_score = (threshold_value - mean_age) / std

from scipy.stats import norm

probability_less_than_80 = norm.cdf(z_score)
probability_over = 1 - probability_less_than_80

# Calculate the number of students expected to score over 80
total_students = 60
expected_students_over_80 = probability_over * total_students

print(f"Expected number of students scoring over 80: {expected_students_over_80:.0f}")
```
