#Data Visualization:
Data visualization involves creating graphical representations of data to make patterns and insights more easily understandable. Common types of visualizations include:

###Histograms:
Histograms are like bar charts that show how numbers are spread out. They help us see which numbers are more common and which are rarer.

###Bar Plots:
Bar plots use bars to compare things, like how many of each thing there are. They're good for showing differences between categories.

###Scatter Plots:
Scatter plots help us understand if two sets of numbers are connected. They use dots to show where points fall on a graph and how they might relate.

###Box Plots:
Box plots help us see the range of numbers and where most values lie. They also show if there are any unusual values, called outliers.


```python
import pandas as pd
import matplotlib.pyplot as plt



# Read student data from CSV
student_data = (pd.read_csv('/content/drive/MyDrive/Data Science/Data Visualization/students.csv'))
#student_data=student_data.head(20)

data=student_data.describe()
print("summary",data)
print()

data1=student_data.info()
print(data1)




# Data Visualization

# Histogram of Ages
plt.figure(figsize=(10, 6))
plt.hist(student_data['Age'], bins=6, color='blue', alpha=0.5)
plt.title('Histogram of Student Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


# Bar chart of Grades
plt.figure(figsize=(10, 6))
plt.bar(student_data['Name'], student_data['GPA'], color='green')
plt.title('Bar Plot of Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=180, va='top')
plt.show()


# Scatter Plot of Age vs. Grade
plt.scatter(student_data['Age'], student_data['GPA'], color='red')
plt.title('Scatter Plot of Age vs. Grade')
plt.xlabel('Age')
plt.ylabel('Grade')
plt.show()

# Box Plot of Grades
plt.boxplot(student_data['GPA'])
plt.title('Box Plot of Student Grades')
plt.ylabel('Grade')
plt.show()
```
